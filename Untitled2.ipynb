```python
import pandas as pd

# Read CSV files
file2 = 'uber-raw-data-apr14.csv'
file3 = 'uber-raw-data-aug14.csv'
file5 = 'uber-raw-data-jul14.csv'
file6 = 'uber-raw-data-jun14.csv'
file7 = 'uber-raw-data-may14.csv'
file8 = 'uber-raw-data-sep14.csv'
```

```python
df2 = pd.read_csv(file3)
df2
```

```python
df3 = pd.read_csv(file3)
df3
```

```python
df5 = pd.read_csv(file5)
df5
```

```python
df6 = pd.read_csv(file6)
df6
```

```python
df7 = pd.read_csv(file7)
df7
```

```python
df8 = pd.read_csv(file8)
df8
```

```python
# Concatenate DataFrames
combined_df = pd.concat([df2, df3, df5, df6, df7, df8], ignore_index=True)
rows_with_null = combined_df.isnull().any(axis=1).sum()
rows_with_null
```

```python
combined_df
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8)  # Specify the number of clusters
combined_df['cluster'] = kmeans.fit_predict(combined_df[['Lat', 'Lon']])
```

```python
import matplotlib.pyplot as plt
# Plot the clustered data
plt.scatter(combined_df['Lon'], combined_df['Lat'], c=combined_df['cluster'], cmap='viridis', marker='o')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clusters of Latitude and Longitude Coordinates')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
for index, row in combined_df.iterrows():
  time = str(row['Date/Time']).split(" ")[1]
  combined_df.at[index, 'time_index'] = time.split(":")[0]
combined_df
```

```python
combined_df = combined_df.sort_values(by='Date/Time')
combined_df
```

```python
cluster_counts = combined_df['cluster'].value_counts()
max_cluster_num = cluster_counts.idxmax()

max_cluster_df = combined_df[combined_df['cluster'] == max_cluster_num]

time_counts = max_cluster_df['time_index'].value_counts()
top_5_time_indexes = time_counts.head(5)
print("Cluster zone is: ", max_cluster_num)
print('\n')
print("Top 5 hours that are peak are: \n", top_5_time_indexes)


```
